```python
# Tratando valores nulos
for coluna in df.columns:
    if (str(df[coluna].dtypes)) == 'float64' or (str(df[coluna].dtypes)) == 'int64':
        df[coluna].fillna(df[coluna].mean(), inplace = True)
    elif df[coluna].dtype == 'datetime64[ns]':
        df[coluna].fillna(pd.NaT, inplace=True)
    else:
        df[coluna].fillna("Ausente", inplace = True)
```

```python
# Transformando em categoricas
df['product_category_name'] = df['product_category_name'].astype('category')
df['product_category_name_english'] = df['product_category_name_english'].astype('category')
df['payment_type'] = df['payment_type'].astype('category')
df['product_category_name'] = df['product_category_name'].astype('category')
df['customer_city'] = df['customer_city'].astype('category')
df['customer_state'] = df['customer_state'].astype('category')
df['seller_city'] = df['seller_city'].astype('category')
df['seller_state'] = df['seller_state'].astype('category')
```

```python
# Identificador de outliers
def drop_outliers(base, variavel):
    outliers = []
    # Calcula Q1 (25º percentil) para a variável dada
    Q1 = np.percentile(base[variavel], 25)
    
    # Calcula Q3 (75º percentil) para a variável dada
    Q3 = np.percentile(base[variavel], 75)
    
    # Usa IQR (interquartile range) para calcular um passo de outlier
    IQR = Q3-Q1
    passo = 1.5 * IQR
    
    feature_outliers = base[~((base[variavel] >= Q1 - passo) & (base[variavel] <= Q3 + passo))]
    outliers.extend(list(feature_outliers.index.values))
    
    return outliers
```

```python
# Retirando todos os outliers dos valores numericos

for coluna in df.columns:
    if (str(df[coluna].dtypes)) == 'float64' or (str(df[coluna].dtypes)) == 'int64':
        outliers = drop_outliers(df, str(coluna))
        dfSemOutliers = df.drop(df.index[outliers]).reset_index(drop = True)

df = dfSemOutliers
```

#### Criando uma coluna que mostra o tempo que o pedido demorou a chegar na transportadora após ser aprovado e outra que mostra o tempo que levou para o pedido ser aprovado

```python
# Função para calcular a diferença entre datas e horários completos
def calcular_tempo_entrega_transportadora(row):
    diferenca_completa = row['order_delivered_carrier_date'] - row['order_approved_at']
    diferenca_em_dias = diferenca_completa.total_seconds() / (24 * 3600)
    return diferenca_em_dias

# Aplicando a função a cada linha do DataFrame
df['tempo_entrega_transportadora'] = df.apply(calcular_tempo_entrega_transportadora, axis=1)
```

```python
# Função para calcular a diferença entre datas e horários completos
def calcular_tempo_aceitar_pedido(row):
    diferenca_completa = row['order_approved_at'] - row['order_purchase_timestamp']
    diferenca_em_dias = diferenca_completa.total_seconds() / (24 * 3600)
    return diferenca_em_dias

# Aplicando a função a cada linha do DataFrame
df['tempo_aceitar_pedido'] = df.apply(calcular_tempo_aceitar_pedido, axis=1)

df
```

```python
# Identificador de outliers
def drop_outliers(base, variavel):
    outliers = []
    # Calcula Q1 (25º percentil) para a variável dada
    Q1 = np.percentile(base[variavel], 25)
    
    # Calcula Q3 (75º percentil) para a variável dada
    Q3 = np.percentile(base[variavel], 75)
    
    # Usa IQR (interquartile range) para calcular um passo de outlier
    IQR = Q3-Q1
    passo = 1.5 * IQR
    
    feature_outliers = base[~((base[variavel] >= Q1 - passo) & (base[variavel] <= Q3 + passo))]
    outliers.extend(list(feature_outliers.index.values))
    
    return outliers
```

```python
# Retirando todos os outliers dos valores numericos

for coluna in df.columns:
    if (str(df[coluna].dtypes)) == 'float64' or (str(df[coluna].dtypes)) == 'int64':
        outliers = drop_outliers(df, str(coluna))
        dfSemOutliers = df.drop(df.index[outliers]).reset_index(drop = True)

df = dfSemOutliers
```

```python
# Convertendo os valores negativos da review para positvos
df['review_score'] = df['review_score'].abs()
```

```python
# Deletando as linhas nas quais o status do pedido é classificado como cancelado pois existem apenas 6 linhas assim
df = df[df['order_status'] != 'canceled']
```

```python
# Calculando o tempo levado para o pedido ser aceito, o tempo levado para o pedido chegar na transportadora após aceito 
# e o tempo levado pro pedido sair da transportadora e chegar no cliente

# Criando funções para calcular os tempos
def calcularTempoEntregaTransportadora(row):
    diferencaCompleta = row['order_delivered_carrier_date'] - row['order_approved_at']
    diferencaEmDias = diferencaCompleta.total_seconds() / (24 * 3600)
    return diferencaEmDias

def calcularTempoAceitarPedido(row):
    diferencaCompleta = row['order_approved_at'] - row['order_purchase_timestamp']
    diferencaEmDias = diferencaCompleta.total_seconds() / (24 * 3600)
    return diferencaEmDias

def calcularTempoEntregaCliente(row):
    diferencaCompleta = row['order_delivered_customer_date'] - row['order_delivered_carrier_date']
    diferencaEmDias = diferencaCompleta.total_seconds() / (24 * 3600)
    return diferencaEmDias

# Fazendo uma cópia do DataFrame original para evitar modificações na visualização
df = df.copy()  

# Aplicando as funções a cada linha do DataFrame usando loc
df.loc[:, 'tempo_entrega_transportadora'] = df.apply(calcularTempoEntregaTransportadora, axis=1)
df.loc[:, 'tempo_aceitar_pedido'] = df.apply(calcularTempoAceitarPedido, axis=1)
df.loc[:, 'tempo_entrega_cliente'] = df.apply(calcularTempoEntregaCliente, axis=1)
```
